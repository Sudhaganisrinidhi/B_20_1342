```python
import matplotlib.pyplot as plt
# Sample data
students = ["A", "B", "C", "D", "E"]
marks = [85, 70, 90, 60, 75]
# Bar chart
plt.bar(students, marks, color="pink")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Performance")
plt.show()
```

```python
#Example: Line Plot of Student Marks
import matplotlib.pyplot as plt
# Data
subjects = ["Math", "Physics", "Chemistry", "Biology",
"English"]
marks = [85, 90, 78, 88, 76]
# Plot
plt.plot(subjects, marks, marker='o', color='green')
# Customization
plt.title("Student Marks in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.grid(True)
# Show plot
plt.show()



```

```python
#Example: Daily Temperature Monitoring
# Data
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperature = [30, 32, 31, 29, 33, 35, 34]
# Plot
plt.plot(days, temperature, marker='s', linestyle='--',
color='blue')
# Customization
plt.title("Weekly Temperature Trend")
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.grid(True)
# Show plot
plt.show()




```

```python
#6. Real-Time Scenario Example

import matplotlib.pyplot as plt
# Data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [15000, 18000, 12000, 22000, 24000, 20000]
# Plot
plt.bar(months, sales, color='red')
# Customization
plt.title("Monthly Sales Report")
plt.xlabel("Months")
plt.ylabel("Sales in USD")
plt.grid(axis='y')
# Show plot
plt.show()



```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=np.random.rand(6,6)
corr=pd.DataFrame(data,columns=list("ABCDEF")).corr()
#plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,cmap="coolwarm",linewidth=0.5)
plt.title("Heatmap Example-Correlation Matrix")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset("iris")
sns.pairplot(df,hue="species",diag_kind="kde",palette="Set2")
plt.suptitle("pair plot Example-iris Dataset",y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
#Barplot
plt.figure(figsize=(6,4))
sns.barplot(x="day",y="total_bill",data=tips,palette="muted")
plt.title("Barplot Example")
plt.show()
#Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(x="day",y="total_bill",data=tips,palette="Set2")
plt.title("Boxplot Example")
plt.show()
#Violinplot
plt.figure(figsize=(6,4))
sns.violinplot(x="day",y="total_bill",data=tips,palette="Set3")
plt.title("Violinplot Example")
plt.show()
#Countplot
plt.figure(figsize=(6,4))
sns.countplot(x="day",data=tips,palette="coolwarm")
plt.title("Countplot Example")
plt.show()
```
