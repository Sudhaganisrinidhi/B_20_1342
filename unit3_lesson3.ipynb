```python
import plotly.express as px
# Sample data
data = {
"Year": [2018, 2019, 2020, 2021, 2022],
"Sales": [100, 150, 130, 170, 200]
}
# Create interactive line chart
fig = px.line(data, x="Year", y="Sales", title="Sales GrowthOver Years", markers=True)
fig.show()
```

```python
import plotly.express as px
data = {
"Products": ["A", "B", "C", "D"],
"Revenue": [40000, 25000, 30000, 50000]
}
fig = px.bar(data, x="Products", y="Revenue", color="Products",
title="Revenue by Product")
fig.show()
```

```python
import plotly.express as px
df = px.data.iris() # Built-in dataset
fig = px.scatter(df, x="sepal_width", y="sepal_length",
color="species",
title="Iris Flower Dataset - Scatter Plot",
size="petal_length", hover_data=["petal_width"])
fig.show()
```

```python
import plotly.express as px
data = {
"Category": ["Electronics", "Clothing", "Groceries", "Toys"],
"Sales": [45000, 25000, 15000, 10000]
}
fig = px.pie(data, values="Sales", names="Category", title="Sales Distribution by Category")
fig.show()
```

```python
import plotly.express as px
df = px.data.tips() # Built-in dataset
fig = px.histogram(df, x="total_bill", nbins=20,
title="Distribution of Total Bill Amounts")
fig.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure()
# Add line trace
fig.add_trace(go.Scatter(x=[1,2,3,4,5], y=[10,15,13,18,20],
mode='lines+markers',
name='Sales'))
# Add bar trace
fig.add_trace(go.Bar(x=[1,2,3,4,5], y=[8,12,10,15,18],
name='Revenue'))
fig.update_layout(title="Sales and Revenue Comparison",
xaxis_title="Month", yaxis_title="Value")
fig.show()
```

```python
from dash import Dash, dcc, html
import plotly.express as px
import pandas as pd
# Create app
app = Dash(__name__)
# Sample data
df = px.data.gapminder().query("country=='India'")
fig = px.line(df, x="year", y="gdpPercap", title="GDP per Capita of India Over Years")
# Layout of dashboard
app.layout = html.Div([
html.H1("Interactive Dashboard Example"),
dcc.Graph(figure=fig)
])
```

```python
import plotly.graph_objects as go
import pandas as pd
import time
fig = go.FigureWidget()
fig.add_scatter(name="Live Data", x=(), y=())
fig.update_layout(title="Live Data Simulation",
xaxis_title="Time", yaxis_title="Value")
for i in range(10):
    fig.data[0].x += (i,)
    fig.data[0].y += (i**2,)
    time.sleep(0.5)
fig.show()
```
