# Average Monthly Rainfall (1000 rows)


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("traffic_source_1000.csv")
```

```python
traffic_summary = df.groupby("Source")["Visits"].sum().reset_index()
```

```python
plt.figure(figsize=(8,5))
bars = plt.bar(traffic_summary['Source'], traffic_summary['Visits'],
               color='skyblue', edgecolor='black')
```

```python
# for bar in bars:
#     plt.text(bar.get_x() + bar.get_width()/2,
#              bar.get_height() + 50,
#              str(bar.get_height()),
#              ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
plt.xlabel("Traffic Source")
plt.ylabel("Total Visits")
plt.title("Website Traffic by Source (Aggregated from 1000 Records)")
plt.tight_layout()
plt.show()
```

```python
highest_source = traffic_summary.loc[traffic_summary['Visits'].idxmax(), 'Source']
print(f" Insight: The highest traffic source is '{highest_source}'.")
```

# Average Monthly Rainfall (1000 rows)


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("monthly_rainfall_1000.csv")
```

```python
rainfall_summary = df.groupby("Month")["Rainfall"].mean().reset_index()

```

```python
season_colors = {
    'Jan':'lightblue','Feb':'lightblue','Dec':'lightblue',
    'Mar':'orange','Apr':'orange','May':'orange',
    'Jun':'green','Jul':'green','Aug':'green','Sep':'green',
    'Oct':'brown','Nov':'brown'
}
```

```python
colors = [season_colors[month] for month in rainfall_summary['Month']]
```

```python
plt.figure(figsize=(10,6))
bars = plt.bar(rainfall_summary['Month'], rainfall_summary['Rainfall'],
               color=colors, edgecolor='black')
```

```python

# Labels and title
plt.xlabel("Month")
plt.ylabel("Average Rainfall (mm)")
plt.title("Average Monthly Rainfall (Aggregated from 1000 Records)")
plt.tight_layout()
plt.show()

# Insights
wettest = rainfall_summary.loc[rainfall_summary['Rainfall'].idxmax(), 'Month']
driest = rainfall_summary.loc[rainfall_summary['Rainfall'].idxmin(), 'Month']
print(f" Insight: The wettest month is '{wettest}', and the driest month is'{driest}'.")
```
