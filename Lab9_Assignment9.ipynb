Lab 9: Visualize Relationships Between Variables Using Seaborn Pair Plots

 Wine Quality Assessment

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# 1. Load the WineQT.csv dataset
file_path = "WineQT.csv"
wine_df = pd.read_csv(file_path)
print("\n--- Wine Dataset Preview ---")
print(wine_df.head())
print("\nDataset Shape:", wine_df.shape)
```

```python
wine_df.columns = [c.strip().lower().replace(" ", "_") for c in wine_df.columns]
```

```python
selected_features = [
   "fixed_acidity",
   "volatile_acidity",
   "citric_acid",
   "residual_sugar",
   "alcohol",
   "quality"
]
data = wine_df[selected_features].dropna()
```

```python
sns.set(style="whitegrid", context="talk")
pair_plot = sns.pairplot(
    data,
    hue="quality",
    diag_kind="kde",
    palette="coolwarm"
)
pair_plot.fig.suptitle("Scenario 6: Relationship Between Wine Properties and Quality", fontsize=14, y=1.03)
plt.show()
```

Student Performance Visualization

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# 1. Create sample dataset
np.random.seed(42)
data = {
     "Study_Hours": np.random.normal(5, 1.5, 50).clip(1, 10),
     "Attendance(%)": np.random.normal(80, 10, 50).clip(50, 100),
     "Exam_Score": np.random.normal(70, 12, 50).clip(30, 100),
     "Gender": np.random.choice(["Male", "Female"], 50)
}
students_df = pd.DataFrame(data)
```

```python
print("\n Student Dataset Preview ")
print(students_df.head())
```

```python
sns.set(style="ticks", context="talk")
pair_plot = sns.pairplot(
    students_df,
    hue="Gender",
    diag_kind="kde",
    palette="Set2"
)
pair_plot.fig.suptitle("Scenario 5: Student Performance Visualization", fontsize=14, y=1.03)
plt.show()
```
